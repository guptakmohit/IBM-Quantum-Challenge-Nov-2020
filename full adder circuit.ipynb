```python
from qiskit import *
```

```python
q = QuantumRegister(8)
c = ClassicalRegister(2)
qc = QuantumCircuit(q,c)
```

```python
qc.draw()
```

```python
qc.x(q[0])
qc.x(q[2])
qc.draw()
```

```python
qc.cx(q[1],q[3])
qc.cx(q[0],q[3])
qc.draw()
```

```python
qc.barrier()
```

```python
qc.ccx(q[0], q[1], q[4])
qc.draw()
```

```python
qc.cx(q[2],q[5])
qc.cx(q[3],q[5])
qc.draw()
```

```python
qc.ccx(q[2],q[3],q[6])
qc.barrier()
qc.draw()
```

```python
qc.cx(q[4],q[7])
qc.cx(q[6],q[7])
qc.ccx(q[4],q[6],q[7])
qc.barrier()
qc.draw()
```

```python
qc.measure(q[5],c[0])
qc.measure(q[7], c[1])
qc.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc, backend, shots=1000)
result = job.result()
count =result.get_counts()
print(count)
qc.draw()
```
